```python
data_list = [json.loads(line) for line in open("kjw2_data/江启臣_20210806.json", encoding='utf8')]
data_list.extend([json.loads(line) for line in open("kjw2_data/张亚中_20210806.json", encoding='utf8')])
data_list.extend([json.loads(line) for line in open("kjw2_data/朱立伦_20210806.json", encoding='utf8')])

data = pd.DataFrame(data_list)
```

```python
pattern = r'(#.*?\s+?)'

all_hashtags = []

for content in data.content:
    hashtags = re.findall(pattern, content)
    # print(hashtags)
    for ht in hashtags:
        ht = ht.strip() 
        if ht.startswith("#江啟臣"):
            ht = "#江啟臣"
        elif ht.startswith("#朱立倫"):
            ht = "#朱立倫"
        elif ht.startswith("#張亞中"):
            ht = "#張亞中"
        all_hashtags.append(ht)

from collections import Counter
all_hashtags_counter = Counter(all_hashtags)
all_hashtags_counter
```

```python
from preprocess import CustomTweetTokenizer, normalize_hashtags

t = CustomTweetTokenizer(hashtags=["江啟臣", "朱立倫", "張亞中"])

with open("train.txt", "w", encoding="utf8") as f:
    for content in data.content:
        label = None
        if "#江啟臣" in content and "#朱立倫" not in content and "#張亞中" not in content:
            label = 1
        if "#江啟臣" not in content and "#朱立倫" in content and "#張亞中" not in content:
            label = 2
        if "#江啟臣" not in content and "#朱立倫" not in content and "#張亞中" in content:
            label = 3
        if label:
            words = t.tokenize(content)
            remove_labelled_hashtags = []
            for w in words:
                if w in ["#江啟臣", "#朱立倫", "#張亞中"]:
                    w = "#HT"
                remove_labelled_hashtags.append(w)
            f.write(f"{label}\t{' '.join(remove_labelled_hashtags)}\n")

```
