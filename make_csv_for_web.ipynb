```python
def csv_to_web(in_name, out_name):
    """
    With MB
    """
    data = pd.read_csv(in_name).set_index("dt")
    data = data.rename(columns={"0": "Pete Buttigieg", 
                                "1": "Bernie Sanders", 
                                "2": "Elizabeth Warren", 
                                "3": "Joe Biden", 
                                "4": "Others",
                                "5": "Mike Bloomberg"})
    # data.to_csv("data/csv/results-14days.csv")
    data_p = data.div(data.sum(axis=1), axis=0) * 100
    data_p = data_p.round(3)
    data_p.to_csv(out_name)
    

# data = pd.read_csv("disk/results-14days.csv").set_index("dt")

def csv_to_web_V2(in_name, out_name):
    """
    Without MB
    """
    data = pd.read_csv(in_name).set_index("dt")
    data = data[["0", "1", "2", "3", "4"]]
    data = data.rename(columns={"0": "Pete Buttigieg", 
                                "1": "Bernie Sanders", 
                                "2": "Elizabeth Warren", 
                                "3": "Joe Biden", 
                                "4": "Others",
                                "5": "Mike Bloomberg"})
    # data.to_csv("data/csv/results-14days.csv")
    data_p = data.div(data.sum(axis=1), axis=0) * 100
    data_p = data_p.round(1)
    data_p.to_csv(out_name)
```

```python
# for election day

sns.set(style="darkgrid", font_scale=1.5)

def plot_state(csv_file):
    state = csv_file.split("-")[-1][:2]
    print(state)
    data = pd.read_csv("disk/" + csv_file).set_index("state")
    data = data.rename(columns={"0": "Pete Buttigieg", 
                                "1": "Bernie Sanders", 
                                "2": "Elizabeth Warren", 
                                "3": "Joe Biden", 
                                "4": "Others"})
    data_p = data.div(data.sum(axis=1), axis=0) * 100
    data_p = data_p[data_p.index==state]

    data_p.plot(kind='bar', width=0.8)
    plt.xticks(rotation=45)
    plt.xlabel("")
    plt.ylabel("Percent")
    plt.legend(loc=4, bbox_to_anchor=(1.35, 0, 0.25, 0.25))
    # plt.legend(loc=4, bbox_to_anchor=(0.75, 1), ncol=5, frameon=False)
    plt.savefig(f"fig/{csv_file}.png", dpi=300, bbox_inches ='tight')
    # plt.close()
    plt.show()

csv_file = "results-14days-0203-IADay.csv"
plot_state(csv_file)
csv_file = "results-14days-0211-NHDay.csv"
plot_state(csv_file)
csv_file = "results-14days-0222-NVDay.csv"
plot_state(csv_file)

csv_file = "results-culFrom01-0203-IADay.csv"
plot_state(csv_file)
csv_file = "results-culFrom01-0211-NHDay.csv"
plot_state(csv_file)
csv_file = "results-culFrom01-0222-NVDay.csv"
plot_state(csv_file)
```
