```python
rst = pd.read_csv("data/csv/0501-states.csv", usecols=["state", "Biden", "Trump"]).set_index("state")
```

```python
rst.plot(kind="bar", figsize=(20, 5))
plt.ylabel("# of users")
plt.savefig("fig/# of users.png", dpi=300, bbox_inches='tight')
```

```python
rst["Biden_percent"] = rst.Biden / (rst.Biden + rst.Trump) * 100
rst["Trump_percent"] = rst.Trump / (rst.Biden + rst.Trump) * 100
rst = rst[["Biden_percent", "Trump_percent"]]
rst = rst.rename(columns={"Biden_percent": "Biden", "Trump_percent": "Trump"})
```

```python
rst.plot(kind="bar", figsize=(20, 5))
plt.ylabel("Percent per state")
plt.savefig("fig/Percent per state.png", dpi=300, bbox_inches='tight')
```
